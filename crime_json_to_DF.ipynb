```python
from math import radians, sin, cos, sqrt, asin, pi, degrees

def miles_to_degrees(lat, lon, dist):
 
  R = 6372.8/1.60934 # Earth radius in miles
  Dlong = pi/180.*R*cos(radians(lat))
  Dlat = pi/180.*R
 
  return (dist/Dlong,dist/Dlat)
```

```python
def GetCrimeDFfromFile(filename):
  f = open(filename, 'r')
  crime_json = json.loads(f.read())
  crime_list=[]
  for crime in crime_json:
    crime_dict={'EANHotelID': 0,'Assault': 0, 'Burglary': 0, 'Other': 0, 'Theft': 0, 'Vandalism': 0,
           'Arrest': 0, 'Arson': 0, 'Robbery': 0, 'Shooting':0}
    crime_dict['EANHotelID'] = crime['eanid']
    for subcrime in crime['crimes']:
      if  datetime.date(2015, 9, 18) <= datetime.datetime.strptime(subcrime['date'][0:8], "%m/%d/%y").date() <= datetime.date(2015, 9, 21):
        crime_dict[subcrime['type']] = crime_dict.get(subcrime['type'], 0) + 1
    crime_list.append(crime_dict)
  f.close()
  return DataFrame(crime_list)
```
